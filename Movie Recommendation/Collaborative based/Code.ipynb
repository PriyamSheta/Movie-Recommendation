```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('ratings.csv')
```

```python
df.head()
```

```python
mov=pd.read_csv('movies.csv')
```

```python
mov.head()
```

```python
data=pd.merge(df,mov,on='movieId')
```

```python
data.head()
```

```python
import matplotlib.pyplot as pyplot
import seaborn as sns
```

```python
data.groupby('title')['rating'].mean().sort_values(ascending=False).head()
```

```python
data.groupby('title')['rating'].count().sort_values(ascending=False).head()
```

```python
ratings=pd.DataFrame(data.groupby('title')['rating'].mean())
ratings.head()
```

```python
ratings['no. of ratings']=pd.DataFrame(data.groupby('title')['rating'].count())
ratings.head()
```

```python
movie_mat=data.pivot_table(index='userId',columns='title',values='rating')
```

```python
movie_mat.head()
```

```python
ratings.sort_values('no. of ratings',ascending=False).head(10)
```

```python
Forrest_Gump=movie_mat['Forrest Gump (1994)']
```

```python
Forrest_Gump.head()
```

```python
similar_movies=movie_mat.corrwith(Forrest_Gump)
```

```python
corr_movies=pd.DataFrame(similar_movies,columns=['corelation'])
```

```python
corr_movies.head()
```

```python
corr_movies.dropna(inplace=True)
corr_movies.head()
```

```python
corr_movies.sort_values('corelation',ascending=False).head(10)
```

```python
# There is a possibility that these movies are rated only once ot twice/
```

```python
corr_movies=corr_movies.join(ratings['no. of ratings'])
```

```python
corr_movies.head()
```

```python
corr_movies[corr_movies['no. of ratings']>100].sort_values('corelation',ascending=False).head(10)
```
